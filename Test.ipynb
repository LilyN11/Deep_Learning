```python
import random
import torchvision.datasets as datasets
from IPython.display import display

data_path = '.'  

# Uses the train test split from the folders
# Transformer is none for PIL
train_dataset = datasets.OxfordIIITPet(
    root = data_path,
    split = 'trainval',
    download = True,
    target_types = 'category',
    transform = None
)
test_dataset = datasets.OxfordIIITPet(
    root = data_path,
    split = 'test',
    download = True,
    target_types = 'category',
    transform = None
)
```

```python
#sample index, filename, and category name.
def random_sample(dataset, n=5):
    indices = random.sample(range(len(dataset)), n)
    for idx in indices:
        img,target = dataset[idx]
        filename = dataset._images[idx]  # _images is internal
        # map categories
        category_name = dataset.classes[target]
        print(f"Index:{idx}, File: {filename}, Category: {category_name}")
        display(img)

# Display random samples from train
random_sample(train_dataset, n=5)
```
